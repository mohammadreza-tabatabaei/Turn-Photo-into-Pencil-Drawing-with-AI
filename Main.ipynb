```python
import cv2
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('/content/drive/My Drive/Pencil Drawing/1.jpg')
```

```python
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
imgplot = plt.imshow(image)
plt.gray()
plt.show()
print('Pencil Drawing')
```

```python
image_Gray=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
imgplot = plt.imshow(image_Gray)
plt.gray()
plt.show()
print('Pencil Drawing')
```

```python
invert=cv2.bitwise_not(image_Gray)
imgplot = plt.imshow(invert)
plt.gray()
plt.show()
print('Pencil Drawing')
```

```python
blur=cv2.GaussianBlur(invert,(21,21),0)
imgplot = plt.imshow(blur)
plt.gray()
plt.show()
print('Pencil Drawing')
```

```python
Invertedblur=cv2.bitwise_not(blur)
imgplot = plt.imshow(Invertedblur)
plt.gray()
plt.show()
print('Pencil Drawing')
```

```python
Sketch_Filter=cv2.divide(image_Gray,Invertedblur,scale=256)
imgplot = plt.imshow(Sketch_Filter)
plt.gray()
plt.show()
print('Pencil Drawing')
```

```python
def Sketch_Convertor(image):
  grey_Filter=cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
  invert=cv2.bitwise_not(grey_Filter)
  blur=cv2.GaussianBlur(invert,(21,21),0)
  Invertedblur=cv2.bitwise_not(blur)
  Sketch_Filter=cv2.divide(grey_Filter,Invertedblur,scale=256)
  return Sketch_Filter
```

```python
image=cv2.imread('/content/drive/My Drive/Pencil Drawing/1.jpg')
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
Sketch_Filter=Sketch_Convertor(image)
imgplot = plt.imshow(image)
plt.gray()
plt.show()
print('Pencil Drawing')
imgplot = plt.imshow(Sketch_Filter)
plt.gray()
plt.show()
```
